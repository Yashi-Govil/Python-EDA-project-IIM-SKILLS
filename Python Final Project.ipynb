```python
retail_df["Day of the Week"].value_counts()
```

```python
plt.figure(figsize=[7,5])
sns.countplot(data=retail_df, x="Day of the Week",color="chocolate",edgecolor="black",width=0.7)
plt.title("Sales Count by Day of the Week")
plt.xlabel("Day Of The Week")
plt.ylabel("Count")  
# Add horizontal gridlines
plt.grid(axis='y', linestyle='-', linewidth=0.8, alpha=0.4)
plt.show
```

```python
# Key Insights
print("Key Insights:")
print("1. Saturday has the most sales followed by the Sunday & Monday.")
print("2. Sales on the rest days were same.")
print("3. Retailers can adjust staffing or promotions based on busy or slow days for e.g., more staff on weekends; discounts mid-week to boost sales.") 
```

```python
retail_df["Product Category"].value_counts()
```

```python
plt.figure(figsize=[6,6])
data =retail_df["Product Category"].value_counts(normalize = True)

labels = ['Furniture','Electronics','Clothing','Groceries']
ex1 = [0.1,0,0,0]    # extension - by how much points the sector will be exploded
custom_palette = ["#4E79A7", "#F28E2B", "#E15759", "#76B7B2"]
colors = sns.color_palette(custom_palette)

plt.pie(data, labels=labels, startangle=90, explode=ex1, shadow=True, colors=colors,autopct='%1.1f%%')
plt.title("Proportion of sales by Product Category")
plt.legend(title="Product Category")
plt.show
```

```python
# Key Insights
print("Key Insights:")
print("1. The largest slice(as indicated by the percentage) represents the category with the highest share of sales i.e, Furniture")
print("2. The smallest slice reveals the product category contributing the least to overall sales i.e, Groceries")
print("3. High-performing categories may need more stock and less promotion; low-performing may need price drops or ad support.")
```

```python
retail_df["Sales Revenue (USD)"].value_counts()
```

```python
plt.figure(figsize=[8,5])
sns.histplot(data=retail_df, x="Sales Revenue (USD)", bins =40, color="purple").set(title="Distribution of Sales Revenue", ylabel="Count")
plt.xlabel("Sales Revenue(USD)")
plt.show
```

```python
# Key Insights
print("Key Insights:")
print("1. The height of bars shows how many transactions fall within specific revenue ranges.")
print("2. A tall cluster at the lower revenue range indicates that most transactions are low-value sales.")
print("3. The histogram is right-skewed (tail extends to the right), it means a few high-value sales exist, but most are small.")
print("4. A lack of mid-range revenue could suggest opportunities to introduce mid-tier pricing/products to fill the gap.")
```

```python
# Histogram with KDE(Kernel Density Estimate) curve for smoother trend visualization."
plt.figure(figsize=[8,5])
sns.histplot(data=retail_df, x="Sales Revenue (USD)",bins =40, kde=True, color="purple").set(title="Distribution of Sales Revenue",ylabel="Count")
plt.show()
```

```python
# Key Insights
print("Key Insights:")
print("1. The KDE curve is likely right-skewed,(e.g., beyond $20000).There are fewer high-value transactions, but they do exist")
print("Most customers make modest purchases, but there's a small group of high-spending buyers — potentially valuable VIP or bulk customers.")
print("2.The tallest bar and KDE peak likely occur in the lower revenue range.")
print("This indicates The business is generating high transaction volume at lower price points — likely driven by products like groceries or clothing.")
```

```python
plt.figure(figsize=[8,5])
sns.scatterplot(data=retail_df, x="Units Sold", y="Sales Revenue (USD)", color="olive")
plt.xlabel("Units Sold")
plt.ylabel("Sales Revenue (USD)")
plt.title("Units Sold Vs Sales Revenue")
plt.show()
```

```python
# Key Insights
print("Key Insights: ")
print("1. The data points form an upward trend, it indicates a positive correlation — as Units Sold increases, Sales Revenue also increases.")
print("This means selling more units typically drives higher revenue. The slope of the spread can also suggest average revenue per unit.")
print("2. If you notice some points with high revenue but low units sold, it indicate high-priced products.Conversely, high units sold with low revenue might represent low-margin items.")
print("The plot reveals product pricing tiers and helps identify which categories may be high-margin vs. low-margin.")
```

```python
plt.figure(figsize=[10,5])
sns.barplot(data=retail_df,x="Discount Percentage", y="Sales Revenue (USD)", hue="Holiday Effect", palette="Set2", width=0.7)
plt.xlabel("Discount Percentage")
plt.ylabel("Sales Revenue(USD)")
plt.title("Discount% & Sales Revenue$ by Holiday Effect")
plt.show()
```

```python
# Key Insights
print("Key Insights:")
print("1. Higher discounts result in lower revenue, it may suggest:Erosion of margins without enough increase in volume.")
print("Find the optimal discount level that maximizes revenue without harming profitability.")
print("2. Holiday bars are consistently higher, it confirms that holidays boost revenue, possibly due to seasonal demand or promotional campaigns.")
print("3. Uneven bar heights across discount levels may reflect:Inconsistent response to pricing ;Variability in product types or marketing strategies.")
print("May require segmented pricing for different product categories or customer groups.")
```

```python
plt.figure(figsize=[9,5])
sns.lineplot(data=retail_df,x="Units Sold",y="Marketing Spend (USD)", marker="o",ms=4,mec='#2E7D32', mfc='#A5D6A7',color="darkcyan")
plt.title("Marketing Spend based on Units sold")
plt.xlabel("Units Sold")
plt.ylabel("Marketing Spend (USD)")
plt.show()
```

```python
# Key Insights
print("key Insights: ")
print("1. Many entries have Marketing Spend (USD) equal to 0, even when some Units Sold are present.")
print("This creates clusters of data points around zero, suggesting:Sales may often happen without active marketing, possibly due to organic demand or repeat customers.")
print("2. For entries where Marketing Spend is non-zero, there appears to be a positive relationship.")
print("Higher spend often corresponds with more units sold.However, the spread is not tightly clustered, indicating — marketing doesn’t always guarantee high sales.")
```

```python
#Coorelation matrix
# corr function - it stands for Correlation

corr_matrix= retail_df[["Units Sold","Sales Revenue (USD)","Discount Percentage","Marketing Spend (USD)"]].corr()
corr_matrix
```

```python
plt.figure(figsize=[8,5])
sns.heatmap(corr_matrix,annot=True,cmap="PuBuGn")
plt.title("Correlation matrix between variable")
plt.show()
```

```python
# Key Insights
print("Key Insights:")
print("1. Strong Positive Correlation:Units Sold ↔ Sales Revenue (USD): +0.62.")
print("This is the only substantial positive correlation in the dataset.Sales volume is a key revenue driver.")
print("2. Weak/No Correlation:Discount Percentage shows near-zero correlation with:Units Sold: -0.0016; Sales Revenue: -0.066; Marketing Spend: -0.00032")
print("The discount rate has minimal linear impact on either sales or revenue in this dataset. Its effect might be non-linear or context-dependent (e.g., only effective during holidays).")
print("3. Marketing Spend (USD) has almost zero correlation with:Units Sold: -0.0048; Sales Revenue: -0.0027")
print("Either marketing is not being tracked effectively, or it’s not producing measurable results in this dataset.")
```
